```python
import pandas as pd
import os
import re
```

```python
files = os.listdir("data")
files.remove('xlxs')
files
```

```python
df_lists = []
```

```python
for file in files:
    myfile = pd.read_csv(os.path.join("data",file) , encoding = "ISO-8859-1")
    df_lists.append(myfile)
full_df = pd.concat(df_lists , join = 'outer' , ignore_index = True)
```

```python
#full_df["naming"] = full_df[['name' , "Name" , "ï»¿name"]][(full_df['Name'].isnull()) & (full_df["name"].isnull())]
full_df["name"] = full_df[['name' , "Name" , "ï»¿name"]].apply(lambda x : '.'.join(x.dropna().astype(str)) , axis = 1)
```
